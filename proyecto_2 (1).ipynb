Cargue los datos de energía del archivo `assets/Energy Indicators.xls`, que es una lista de indicadores de [suministro de energía y producción de electricidad renovable](assets/Energy%20Indicators.xls) de las [Naciones Unidas](https://unstats.un.org/unsd/environment/excel_file_tables/2013/Energy%20Indicators.xls) para el año 2013, y asigna esos datos a un DataFrame con el nombre de variable **energy**.
Ten en cuenta que este es un archivo de Excel y no un archivo de valores separados por comas (.csv). Además, asegúrate de excluir la información del pie de página y del encabezado del archivo de datos. Las primeras dos columnas son innecesarias, por lo que debes deshacerse de ellas. También debes cambiar las etiquetas de las columnas

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

file_name = '/content/proyecto_2/assets/Energy Indicators.xls'
energy = pd.read_excel (file_name, header = None, skiprows=18, skipfooter= 38)  # Leo el archivo .xls con pd.read_excel. Al leerlo le exigo que no contenga encabezados, que omita las primeras 18 filas y las ultimas 38.

energy = energy.drop (energy.columns [[0,1]], axis =1)  # Utilizo el método drop para eliminar las columnas seleccionadas [0,1], el argumento 'axis=1' indica que se estan eliminando columnas.

columns_names = ['Country', 'Energy Supply', 'Energy supply per Capita', '% Renewable']  # Establezo el nuevo nombre de las columnas.
energy.columns = columns_names  # Agrego los nombres nuevos a cada columna
```

Convierta `Energy Supply` a gigajulios (**Nota: hay 1,000,000 giga julios en un petajulio**). Para todos los países que carezcan de datos (por ejemplo, datos con \"...\"), asegúrese de que estos reflejen valores `np.NaN`.

```python
energy ['Energy Supply'].replace ('...', np.nan, inplace = True)  # Establezco que los elementos en la columna 'Energy Supply' que sean iguales a '...' sean modificados por np.nan, que representa un valor faltante en Pandas. Utilizo 'inplace=True' para que los cambios se realicen en el DataFrame 'energy' directamente.
energy ['Energy Supply'] *= 1000000  # Cada elemento de la columna 'Energy Supply' es multiplicado por 1000000 (1 petajulio = 1,000,000 gigajulio)
```

Cambia el nombre de los siguientes países según listado. También hay varios países que contienen paréntesis y numeros en su nombre. Asegúrete de eliminarlos.

```python
name_change = {
    "Republic of Korea": "South Korea",
    "United States of America20" : "United States",
    "United Kingdom of Great Britain and Northern Ireland19": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
  }  # Diccionrio que contiene los cambios de nombres

energy ['Country'] = energy ['Country'].replace (name_change)  # Especifico que los cambios contenidos en el diccionario name_change se deben realizar en la columna 'Country'
energy = energy.sort_values (by = 'Country', ascending = True)  # Ordeno la columna 'Country' alfabeticamente (para eso sirve el ascending)
energy ['Country'] = energy ['Country'].str.replace (r'\s*\([^)]*\)', '', regex = True )  # Cambio los caracteres que siguen luego de cero o mas especios y un ( hasta un cierre de parentesis por un espacio vacio en cada elemento de la columna 'Country'
energy ['Country'] = energy ['Country'].str.replace(r'\d', '', regex = True)  # Cambio el digito que le sigue al nombre de un país por un espacio vacio
energy.head()
```

¿Cuáles son los 15 países con mayor GDP (PBI) promedio en los últimos 10 años?

*   Esta función debería devolver una serie llamada `avgGDP` con 15 países y su GDP (PBI) promedio ordenado en orden descendente.


```python
def avgGDP (merged_df):
    gdp_accumulated = merged_df.loc[:, '2006':'2015']  # Tomamos el dataframe con el ranking de 15 paises y con '.loc' seleccionamos todas las filas [':'] y las columnas desde '2006' hasta '2015'.
    mean = gdp_accumulated.mean(axis = 1)  # Luego se calcula el promedio  de las columnas seleccionadas para cada fila (en este caso, para cada país).
    avgGDP_sorted = mean.sort_values(ascending=True)  # Se ordena la serie en orden descendente

    return avgGDP_sorted

average_gdp = avgGDP (merged_df)
average_gdp
```

¿Cuánto cambió el GDP (PBI) durante el lapso de 10 años para el país con el sexto GDP (PBI) promedio más grande?

*   Esta función debería devolver un solo número.



```python
def change_of_gdp (average_gdp, merged_df):
      country_sixth_avgGDP = average_gdp.index[-6]
      gdp_2006 = merged_df.loc[country_sixth_avgGDP, '2006']
      gdp_2015 = merged_df.loc[country_sixth_avgGDP, '2015']
      change_gdp = gdp_2015 - gdp_2006

      return f"El cambio en el PBI durante 10 años para {country_sixth_avgGDP} fue de {change_gdp} puntos."

cambio_pbi = change_of_gdp (average_gdp, merged_df)
cambio_pbi
```

¿Qué país tiene el máximo `% Renewable` y cuál es el porcentaje?

*   Esta función debe devolver una tupla con el nombre del país y el porcentaje



```python
def max_renewable (merged_df):
    df = merged_df [merged_df ['% Renewable'] != 0]
    max_percentage = df ['% Renewable'].max()
    max_index = df ['% Renewable'].idxmax()
    return max_index, max_percentage

max_renewable (merged_df)
```

Cree una nueva columna que sea la relación entre `Self-Citations` y `Total Citations`.
¿Cuál es el valor máximo para esta nueva columna y qué país tiene la relación mas alta?

*   Esta función debe devolver una tupla con el nombre del país y el ratio (proporción).

```python
def relation_columns (merged_df):
    merged_df ['Citation ratio'] = merged_df ['Self-citations'] / merged_df ['Citations']
    max_ratio = merged_df ['Citation ratio'].max ()
    max_index = merged_df ['Citation ratio'].idxmax()

    return max_index, max_ratio

relation_columns (merged_df)
```

Crea una columna que estime la población utilizando `Energy Supply` y `Energy Supply per capita`.
¿Cuál es el tercer país más poblado según esta estimación?

*   Esta función debería devolver el nombre del país



```python
def population_density (merged_df):
    merged_df ['Population'] = merged_df ['Energy Supply'] / merged_df ['Energy supply per Capita']
    merged_df = merged_df.sort_values (by = 'Population', ascending = False)
    third_country = merged_df ['Population'].index[2]

    return third_country

population_density (merged_df)
```

Crea una nueva columna con un valor igual a 1 si el valor de % de energía renovable del país (`% Renewable`) es igual o superior de la mediana de todos los países en el top 15, o con un valor igual a 0 si el valor de % de energía renovable del país (`% Renewable`) está por debajo de la mediana.

*   Esta función debería devolver una serie llamada `HighRenew` cuyo índice es el nombre del país ordenado ascendentemente según el ranking.



```python
def new_column (merged_df):
    median_renewable= merged_df ['% Renewable'].median()  # Mediana de la columna '% Renewable'
    merged_df ['High Renew'] = merged_df ['% Renewable'] >= median_renewable  # True o False
    high_renew = merged_df ['High Renew'].astype(int)  # Pasamos los valores booleanos a 0 y 1 (int) y los guardamos en la variable
    return high_renew

new_column (merged_df)
```

Utiliza el siguiente diccionario para agrupar los países por continente, luego crea un DataFrame que muestre el tamaño de la muestra (la cantidad de países en cada contenedor de continente) y la suma, la media y la desviación estándar para la población estimada de cada país.



*  Esta función debería devolver un DataFrame con un índice llamado Continent `['Asia', 'Australia', 'Europe', 'North America', 'South America']` y columnas `['size', 'sum', 'mean', 'std']`


```python
import pandas as pd

# Diccionario de continentes
continent_dict = {
    'China': 'Asia',
    'United States': 'North America',
    'Japan': 'Asia',
    'United Kingdom': 'Europe',
    'Russian Federation': 'Europe',
    'Canada': 'North America',
    'Germany': 'Europe',
    'India': 'Asia',
    'France': 'Europe',
    'South Korea': 'Asia',
    'Italy': 'Europe',
    'Spain': 'Europe',
    'Iran': 'Asia',
    'Australia': 'Australia',
    'Brazil': 'South America'
}

merged_df['Continent'] = merged_df.index.map(continent_dict)  # Agrupo por continente y sumo la población de cada pais del continente
continent_summary = merged_df.groupby('Continent')['Population'].agg(size='count', sum='sum', mean = 'mean', std= 'std').reset_index()  # El método .agg() en Pandas se utiliza para realizar agregaciones en un DataFrame, osea que calcula una o más estadísticas sobre los datos en una o varias columnas.
continent_summary.set_index('Continent', inplace=True)  # Establecemos la columna 'Continent' como indice del dataframe
continent_summary.columns = ['Size', 'Sum', 'Mean', 'Std']  # Se renombran las columnas con los nombres indicados

continent_summary
```
